### imdb intro
https://www.imdb.com/search/title/?count=100&groups=top_1000&sort=user_rating

```python
# pip install selenium
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import pandas as pd

#Initializing the webdriver
PATH = "C:\Program Files (x86)\chromedriver.exe"
options = Options()
options.add_argument("start-maximized")
options.add_experimental_option("excludeSwitches", ["enable-automation"])
options.add_experimental_option('useAutomationExtension', False)
driver = webdriver.Chrome(chrome_options=options, executable_path=PATH)

#Open Specified URL with keyword you want to search for
URL = "https://www.imdb.com/search/title/?count=100&groups=top_1000&sort=user_rating"
driver.get(URL)

num_movies = 250
movies = []

while len(movies) < num_movies:
    movie_button = driver.find_elements_by_xpath('.//div[@class="lister-item mode-advanced"]')
    
    for movie in movie_button:

        print("Progress: {}".format("" + str(len(movies)) + "/" + str(num_movies)))
        if len(movies) >= num_movies:
            break

        title = movie.find_element_by_xpath('.//h3[@class="lister-item-header"]').text.split('.')[1].strip()
        duration = movie.find_element_by_xpath('.//span[@class="runtime"]').text
        genre = movie.find_element_by_xpath('.//span[@class="genre"]').text
        rate = movie.find_element_by_xpath('.//strong').text  
        descr = movie.find_element_by_xpath('.//p[@class="text-muted"]').text
        cast = movie.find_element_by_xpath('.//p[@class=""]').text
        info = movie.find_element_by_xpath('.//p[@class="sort-num_votes-visible"]').text

        movies.append({
            "Title": title,
            "Duration": duration,
            "Genre": genre,
            "Rate": rate,
            "Description": descr,
            "Cast": cast,
            "Info": info
            })
    if len(movies) >= num_movies:
        print("Scraping terminated before reaching target number of movies. Needed {}, got {}.".format(num_movies, len(movies)))
#         driver.quit()
        driver.close()
        break
        
    driver.find_element_by_xpath('.//a[@class="lister-page-next next-page"]').click()

df = pd.DataFrame(data=movies)
display(df)
```

### imdb try-except
https://www.imdb.com/search/title/?count=100&groups=top_1000&sort=user_rating

```python
# pip install selenium
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException, StaleElementReferenceException
import pandas as pd
import numpy as np
import time


def get_movies(num_movies, path):
    
    #Initializing the webdriver
    options = Options()
    options.add_argument("start-maximized")
    options.add_experimental_option("excludeSwitches", ["enable-automation"])
    options.add_experimental_option('useAutomationExtension', False)
    driver = webdriver.Chrome(chrome_options=options, executable_path=path)

    driver.set_window_size(1120, 1000) #https://stackoverflow.com/questions/15397483/how-do-i-set-browser-width-and-height-in-selenium-webdriver
    

    #Open Specified URL with keyword you want to search for
    URL = "https://www.imdb.com/search/title/?count=100&groups=top_1000&sort=user_rating"
    driver.get(URL)
    
    movies = []
    ignored_exceptions=(NoSuchElementException,StaleElementReferenceException,)
    

    while len(movies) < num_movies:
        
        try:
            #Wait until the web page load
            # https://selenium-python.readthedocs.io/waits.html
            element = WebDriverWait(driver, 10, ignored_exceptions=ignored_exceptions).until(
                    EC.presence_of_element_located((By.ID, "main"))
                    )
        
            movie_button = element.find_elements_by_xpath('.//tbody[@class="list"]')

            for movie in movie_button:

                print("Progress: {}".format("" + str(len(movies)) + "/" + str(num_movies)))
                if len(movies) >= num_movies:
                    break

                try:
                    title = movie.find_element_by_xpath('.//h3[@class="lister-item-header"]').text.split('.')[1].strip()
                except NoSuchElementException:
                    title = np.NaN
                
                try: 
                    certificate = movie.find_element_by_xpath('.//span[@class="certificate"]').text
                except NoSuchElementException:
                    certificate = np.NaN
                    
                try:    
                    duration = movie.find_element_by_xpath('.//span[@class="runtime"]').text
                except NoSuchElementException:
                    duration = np.NAN
                    
                try:
                    genre = movie.find_element_by_xpath('.//span[@class="genre"]').text
                except NoSuchElementException:
                    genre = np.NaN
                    
                try:
                    rate = movie.find_element_by_xpath('.//strong').text 
                except NoSuchElementException:
                    rate = np.NaN
                
                try: 
                    metascore = movie.find_element_by_xpath('.//span[@class="metascore  favorable"]').text
                except NoSuchElementException:
                    metascore = np.NaN
                    
                try:
                    descr = movie.find_element_by_xpath('.//p[@class="text-muted"]').text
                except NoSuchElementException:
                    descr = np.NaN
                
                try:
                    cast = movie.find_element_by_xpath('.//p[@class=""]').text
                except NoSuchElementException:
                    cast = np.NaN
                
                try:
                    info = movie.find_element_by_xpath('.//p[@class="sort-num_votes-visible"]').text
                except NoSuchElementException:
                    info = np.NaN
                
                movies.append({
                    "Title": title,
                    "Certificate": certificate,
                    "Duration": duration,
                    "Genre": genre,
                    "Rate": rate,
                    "Metascore": metascore,
                    "Description": descr,
                    "Cast": cast,
                    "Info": info
                    })
            try:
                if len(movies) < num_movies:
                    element.find_element_by_xpath('.//a[@class="lister-page-next next-page"]').click()
                    time.sleep(20)
                
            except NoSuchElementException:
                print("Scraping terminated before reaching target number of movies. Needed {}, got {}.".format(num_movies, len(movies)))
                driver.quit()
                break   
            
        finally:
            if len(movies) >= num_movies:
                driver.quit()

    return pd.DataFrame(data=movies)

PATH = "C:\Program Files (x86)\chromedriver.exe"
df = get_movies(10, PATH)
        
# df.to_csv("Data/IMDB top 1000.csv")
display(df)
```

### sofifa
https://sofifa.com/

```python
# pip install selenium
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException, StaleElementReferenceException,TimeoutException
import pandas as pd
import numpy as np
import time

def get_data(num_players, path):
    
    #Initializing the webdriver
    options = Options()
    options.add_argument("start-maximized")
    options.add_experimental_option("excludeSwitches", ["enable-automation"])
    options.add_experimental_option('useAutomationExtension', False)
    driver = webdriver.Chrome(chrome_options=options, executable_path=path)

    driver.set_window_size(1120, 1000)
    
    URL = "https://sofifa.com/?offset="
    driver.get(URL)
    
    players = []
    ignored_exceptions=(NoSuchElementException,StaleElementReferenceException, TimeoutException)
    
    while len(players) < num_players:
        
        players_xpaths = driver.find_elements_by_xpath('.//tbody[@class="list"]/tr')

        for player in players_xpaths:

            print("Progress: {}".format("" + str(len(players)) + "/" + str(num_players)))
            if len(players) >= num_players:
                break

            try:
                name = player.find_element_by_xpath('.//div[@class="bp3-text-overflow-ellipsis"]').text
            except NoSuchElementException:
                name = np.nan
                
            try:
                best_position = player.find_element_by_xpath('.//a[@rel="nofollow"]/span').text
            except NoSuchElementException:
                best_position = np.nan

            try: 
                age = player.find_element_by_xpath('.//td[@class="col col-ae"]').text
            except NoSuchElementException:
                age = np.nan

            try:    
                overall_rating = player.find_element_by_xpath('.//td[@class="col col-oa"]/span').text
            except NoSuchElementException:
                overall_rating = np.nan

            try:
                potential = player.find_element_by_xpath('.//td[@class="col col-pt"]/span').text
            except NoSuchElementException:
                potential = np.nan

            try:
                value = player.find_element_by_xpath('.//td[@class="col col-vl"]').text 
            except NoSuchElementException:
                value = np.nan

            try: 
                wage = player.find_element_by_xpath('.//td[@class="col col-wg"]').text 
            except NoSuchElementException:
                wage = np.nan
            
            try: 
                total_stats = player.find_element_by_xpath('.//td[@class="col col-tt"]/span').text 
            except NoSuchElementException:
                total_stats = np.nan
                
            try:
                hits = player.find_element_by_xpath('.//td[@class="col-comment"]').text 
            except NoSuchElementException:
                hits = np.nan

            try:
                team = player.find_element_by_xpath('.//div[@class="bp3-text-overflow-ellipsis"]/a').text 
            except NoSuchElementException:
                team = np.nan

            try:
                contract = player.find_element_by_xpath('.//div[@class="bp3-text-overflow-ellipsis"]/div').text 
            except NoSuchElementException:
                contract = np.nan

            players.append({
                "name": name,
                "best_position" : best_position,
                "age": age,
                "overall_rating": overall_rating,
                "potential": potential,
                "value": value,
                "wage": wage,
                "total_stats":total_stats,
                "hits": hits,
                "team": team,
                "contract": contract
                })
        try:
            if len(players) < num_players:
                driver.find_element_by_xpath('.//a[@class="bp3-button bp3-intent-primary pjax"]').click()
                time.sleep(20)

        except NoSuchElementException:
            print("Scraping terminated before reaching target number of players. Needed {}, got {}.".format(num_players, len(players)))
            driver.quit()
            break   

    return pd.DataFrame(data=players)

PATH = "C:\Program Files (x86)\chromedriver.exe"
df = get_data(10, PATH)

display(df)
```
